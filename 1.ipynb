NEXT PART IS DONE BY REMOVING THE OUTLIERS FROM THE COPY CREATED


```python
q = data['writing score'].quantile(0.1) #used to remove the outliers
data_2 = data[data['writing score']>q]
```

```python
sns.distplot(data_2['writing score'])
```

```python
q = data['math score'].quantile(0.1)
data_2 = data[data['math score']>q]
```

```python
sns.distplot(data_2['math score'])
```

```python
q = data['reading score'].quantile(0.1)
data_2 = data[data['reading score']>q]
```

```python
sns.distplot(data_2['reading score'])
```

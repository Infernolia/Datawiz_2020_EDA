```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#importing the dataset
df=pd.read_csv('practice.csv')
```

```python
#first five rows of dataset
df.head()
```

```python
#to check if any null values are present in the dataset
df.info()
```

No null values are present here.

```python
#for checking duplicates in the dataset
df.drop_duplicates()
```

```python
#gives statistical data
df.describe()
```

```python
#describe the distribution of the data by their quartiles
sb.boxplot(data=df[['math score','writing score','reading score']])
```

```python
#eliminating the data points that you consider as outliers(trimming)
index=df[(df['math score']<30) & (df['writing score']<40) & (df['reading score']<40)].index
df.drop(index, inplace=True)
         
```

```python
sb.boxplot(data=df[['math score','writing score','reading score']])
```

```python
#for visualising co-variance
sb.heatmap(df.corr(), annot=True)
```

```python
#to get graphs of covariance
sb.pairplot(data=df)
```

```python
#for comparing math score and reading score(bivariate analysis)
sb.jointplot(x='math score',y='reading score', data=df, alpha=0.7, kind='hex',color='g')
```

```python
#histograms for each
df.hist(['math score', 'reading score', 'writing score'], alpha=0.7, figsize=(5,5))
```

```python
#counts the number of observations in each category
plt.figure(figsize=(10,10))
sb.countplot(x='race/ethnicity',hue='parental level of education', data=df)
```

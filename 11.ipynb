# PIE CHART ON THE RACE OF THE STUDENTS

```python
type_a = practicedata.loc[practicedata['race/ethnicity']=="group A"].count()[0]
type_b = practicedata.loc[practicedata['race/ethnicity']=="group B"].count()[0]
type_c = practicedata.loc[practicedata['race/ethnicity']=="group C"].count()[0]
type_d = practicedata.loc[practicedata['race/ethnicity']=="group D"].count()[0]
type_e = practicedata.loc[practicedata['race/ethnicity']=="group E"].count()[0]
explode = [0.1,0,0.3,0,0]
colors = ['#D4E446','#54EE5C','#EE54E2','#5471EE','#ED7528']
labels = ["type A","type B","type C","type D","type E"]
plt.pie([type_a,type_b,type_c,type_d,type_e],labels = labels,explode = explode,colors = colors,autopct ='%.2f %%')
plt.show()
```

# PIE CHART ON THE EDUCATION LEVEL OF PARENTS

```python
bachelor_degree = practicedata.loc[practicedata['parental level of education'] == "bachelor's degree"].count()[0]
some_college = practicedata.loc[practicedata['parental level of education'] == "some college"].count()[0]
master_degree = practicedata.loc[practicedata['parental level of education'] == "master's degree"].count()[0]
associate_degre = practicedata.loc[practicedata['parental level of education'] == "associate's degree"].count()[0]
highschool = practicedata.loc[practicedata['parental level of education'] == "high school"].count()[0]
some_highschool = practicedata.loc[practicedata['parental level of education'] == "some high school"].count()[0]

labels = ["bachelor's degree","some college","master's degree","associate's degree","high school","some highschool"]
explode = [0,0,0.1,0.2,0,0.1]
plt.pie([bachelor_degree,some_college,master_degree,associate_degree,highschool,some_highschool],labels=labels,explode = explode,autopct='%.2f %%')
plt.rcParams["figure.figsize"] = (15,10)
```

## PIECHART ON TEST PREPARATION COURSE 

```python
none = practicedata.loc[practicedata['test preparation course'] == "none"].count()[0]
completed = practicedata.loc[practicedata['test preparation course'] == "completed"].count()[0]
labels = ['none course taken','courses taken and completed']
explode = [0,0.08]
plt.pie([none,completed], labels = labels,explode=explode,autopct = '%.2f %%')
plt.show()
```

## PIE CHART ON GENDER/SEX

```python
male = practicedata.loc[practicedata['gender'] == 'male'].count()[0]
female = practicedata.loc[practicedata['gender'] == 'female'].count()[0]
labels = ['male','female']
explode = [0,0.06]
colors = ['#52E9E8','#EC5F5F']
plt.pie([male,female], labels=labels,explode = explode, colors = colors,autopct = '%.2f%%')
plt.show()
```

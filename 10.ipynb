```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Practice.csv")
```

```python
df.info()
```

```python
print("The mean score of mathematics is",df['math score'].mean())
```

```python
print("The mean score of reading is",df['reading score'].mean())
```

```python
print("The mean score of writing is",df['writing score'].mean())
```

```python
df.plot(kind='hist', y='math score', bins=10, ec="Black", color="Red", lw=2)
plt.ylabel('Number of students')
plt.xlabel("Score in Maths")
```

```python
df.plot(kind='hist', y='reading score', bins=10, ec="Blue", color="LightBlue", lw=2)
plt.ylabel('Number of students')
plt.xlabel('Score in reading')
```

```python
df.plot(kind='hist', y='writing score', bins=10, ec="purple", color="pink", lw=2)
plt.ylabel('Number of students')
plt.xlabel('Score in writing')
```

```python
df["Total"] = df["math score"] + df["writing score"] + df["reading score"]
```

```python
sns.catplot(x='race/ethnicity', y='Total', hue='test preparation course', data=df, kind='bar', palette="Set2")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Total Score")
plt.show()
```

```python
sns.stripplot(x="lunch",y='Total',data=df, palette="Set2")   
plt.xlabel("Lunch")
plt.ylabel("Total Score")
plt.show()
```

```python
#catplot provides a new framework giving access to several types of plots in a single figure. In the above
#figure we can see how many students of a particular race are completing the course and their performance
```

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
practice = 'Practice.csv'
sheet1=pd.read_csv(practice,sep=',')
sheet1.head()
```

```python
sb.pairplot(sheet1,hue = 'gender' ,vars = ['math score','reading score','writing score'])
```

```python
sb.pairplot(sheet1,hue = 'race/ethnicity' ,vars = ['math score','reading score','writing score'])
```

```python
sb.pairplot(sheet1,hue = 'parental level of education' ,vars = ['math score','reading score','writing score'])
```

```python
sb.pairplot(sheet1,hue = 'lunch' ,vars = ['math score','reading score','writing score'])
```

```python
sb.pairplot(sheet1,hue = 'test preparation course' ,vars = ['math score','reading score','writing score'])
```

```python
b = sb.countplot(sheet1['gender'])
b.axes.set_title('Gender Distribution', fontsize = 20)
b.set_xlabel('Gender', fontsize = 20)
b.set_ylabel('Count', fontsize = 20)
plt.show()
b = sb.countplot(sheet1['race/ethnicity'])
b.axes.set_title('Race/ethnicity- Distribution', fontsize = 20)
b.set_xlabel('race/ethnicity', fontsize = 20)
b.set_ylabel('Count', fontsize = 20)
plt.show()
b = sb.countplot(sheet1['parental level of education'])
b.axes.set_title('Level of Education- Distribution', fontsize = 20)
b.set_xlabel('Level of Education', fontsize = 20)
b.set_ylabel('Count', fontsize = 20)
plt.show()
```

```python
b = sb.countplot(sheet1['lunch'])
b.axes.set_title('lunch - Distribution', fontsize = 20)
b.set_xlabel('Lunch Type', fontsize = 20)
b.set_ylabel('Count', fontsize = 20)
plt.show()
```

```python
mks_lst=[i/10 for i in range(1000)]
plt.scatter(sheet1['math score'],mks_lst,label='Marks',s=10,marker='o',color='red')
plt.xlabel('Scores in Math',fontsize=20)
plt.ylabel('Score',fontsize=20)
plt.title('Math Score Variation',fontsize=20)
plt.legend()
plt.show()
plt.scatter(sheet1['reading score'],mks_lst,label='Marks',s=10,marker='o')
plt.xlabel('Scores in Reading',fontsize=20)
plt.ylabel('Score',fontsize=20)
plt.title('Reading Score Variation',fontsize=20)
plt.legend()
plt.show()
plt.scatter(sheet1['writing score'],mks_lst,label='Marks',s=10,marker='o',color='green')
plt.xlabel('Scores in Writing',fontsize=20)
plt.ylabel('Score',fontsize=20)
plt.title('Writing Score Variation',fontsize=20)
plt.legend()
plt.show()
```

```python
slices=[sum(sheet1['math score']),sum(sheet1['reading score']),sum(sheet1['writing score'])]
distri=['Math','Reading','Writing']
cols=['m','g','r']
plt.pie(slices,labels=distri,colors=cols,startangle=90,shadow=True,explode=(0,0.3,0.1),autopct='%1.1f%%')
plt.title('Total Marks Scored',fontsize=20)
plt.show()
```

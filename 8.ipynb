```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
data=pd.read_excel("Practice.xlsx")
```

Histogram Plot

```
plt.hist(data['math score'],bins=50,color='blue')
plt.hist(data['reading score'],bins=50,color='yellow')
plt.hist(data['writing score'],bins=50,color='lime')
plt.xlabel("Math,reading&Writing scores")
plt.ylabel("frequency")
plt.show()
```

Count plot

```
sns.countplot('gender',data=data)
```

```
sns.countplot('lunch',data=data)
```

```
sns.countplot('test preparation course',data=data)
```

Pie Chart

```
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.axis('equal')
labels=['Group A','Group B','Group C','Group D','Group E']
students=['89','190','319','262','140']
ax.pie(students,labels=labels,autopct='%1.2f%%')
plt.show()
```

```
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.axis('equal')
labels=["associate's degree","bachelor's degree","high school","master's degree","some college","some high school"]
people=['222','118','196','59','226','179']
ax.pie(people,labels=labels,autopct='%1.2f%%')
plt.show()
```

#**Correlation btw. Level of education and type of lunch (may depict financial status):**
To study whether education level reflects in a person's financial status or not. These estimates can be done by the basis of lunch charges incurred by the students which can denote if a particular candidate has a financial position above/below a particular level.
A side-by-side comparison is done below.

```

hs=data[data['parental level of education']=='high school']
hs=hs['lunch'].value_counts().reset_index()
hs.columns=["Type","Number"]
shs=data[data['parental level of education']=='some high school']
shs=shs['lunch'].value_counts().reset_index()
shs.columns=["Type","Number"]
sc=data[data['parental level of education']=='some college']
sc=sc['lunch'].value_counts().reset_index()
sc.columns=["Type","Number"]
bs=data[data['parental level of education']=='bachelor\'s degree']
bs=bs['lunch'].value_counts().reset_index()
bs.columns=["Type","Number"]
ms=data[data['parental level of education']=='master\'s degree']
ms=ms['lunch'].value_counts().reset_index()
ms.columns=["Type","Number"]
asd=data[data['parental level of education']=='associate\'s degree']
asd=asd['lunch'].value_counts().reset_index()
asd.columns=["Type","Number"]
s=pd.DataFrame(columns=['parental level of education','free/reduced lunch charges','standard lunch charges'])
s['parental level of education']=['some high school','high school','some college','associate\'s degree',"master's degree","bachelor's degree"]
s["free/reduced lunch charges"]=[int(shs[shs["Type"]=='free/reduced']["Number"]),int(hs[hs["Type"]=='free/reduced']["Number"]),int(sc[sc["Type"]=='free/reduced']["Number"]),int(asd[asd["Type"]=='free/reduced']["Number"]),int(ms[ms["Type"]=='free/reduced']["Number"]),int(bs[bs["Type"]=='free/reduced']["Number"])]
s["standard lunch charges"]=[int(shs[shs["Type"]=='standard']["Number"]),int(hs[hs["Type"]=='standard']["Number"]),int(sc[sc["Type"]=='standard']["Number"]),int(asd[asd["Type"]=='standard']["Number"]),int(ms[ms["Type"]=='standard']["Number"]),int(bs[bs["Type"]=='standard']["Number"])]
s.head() #you may have a look at this derived dataset from the original
s.plot.bar(x='parental level of education',rot=13,width=0.5)
pyplot.title("Effect of Parental Education on Personal Finances")
pyplot.ylabel("No. of students")
```

From the above chart, we get an estimate of the fraction of parents having a particular level of education whose children i.e the students incur the particular lunch charges thereby depicting a relation between education level and financial status.

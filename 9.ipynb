```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv(r"D:\Data Science\Data\Practice.csv")
'''Calculating the mean of the subjects'''

df['Average'] = (df['math score'] + df['reading score'] + df['writing score'])/3

''' Collaborating the 'some high school' & 'high school' label'''

for i in range (0, 999):
    if(df.iloc[i,2] == 'some high school'):
        df.iloc[i,2] = 'high school'
'''Converting the column of dataframe to list using .tolist() function'''

parental_education = df['parental level of education'].tolist()

''' Removing duplication to count the number of occurences'''

number_of_different_parental_education_set = set(parental_education)
number_of_different_parental_education_list = list(number_of_different_parental_education_set)
count_of_parental_education = []
''' The nested for loop is used so as to count the number of occurences'''
for i in number_of_different_parental_education_set:
    count = 0
    for j in parental_education:
        if ( i == j):
            count += 1
    count_of_parental_education.append(count)

''' explode is used to pull out a piece of pie chart to highlight it'''

explode = [0,0,0,0.2,0]
plt.pie(count_of_parental_education, labels = number_of_different_parental_education_list,explode = explode, shadow = True, startangle = 90, autopct = "%1.1f%%", wedgeprops = {'edgecolor' : 'black'})
plt.title('Distribution of students according to Parental Education')
plt.style.use('seaborn-poster')
plt.rcParams["axes.edgecolor"] = "0.15"
plt.rcParams["axes.linewidth"]  = "1.25"
plt.tight_layout()
plt.plot      


```

```python
''' Relationship between Parental Education and type of Lunch(Nutrition) provided '''


df_college = df[df['parental level of education'] == 'some college']
df_high = df[df['parental level of education'] == 'high school']
df_associate = df[df['parental level of education'] == "associate's degree"]
df_bachelors = df[df['parental level of education'] == "bachelor's degree"]
df_masters = df[df['parental level of education'] == "master's degree"]

college_lunch_std = 0
college_lunch_free = 0
college_lunch = df_college['lunch'].tolist()
''' Segregating students on the basis of Free and Standard lunch '''
for i in college_lunch :
    if ( i == 'free/reduced'):
        college_lunch_free += 1
    else:
        college_lunch_std += 1
    
college_lunch_total = (college_lunch_free + college_lunch_std)/100

high_lunch_std = 0
high_lunch_free = 0
high_lunch = df_high['lunch'].tolist()
for i in high_lunch :
    if ( i == 'free/reduced'):
        high_lunch_free += 1
    else:
        high_lunch_std += 1
    
high_lunch_total = (high_lunch_free + high_lunch_std)/100

associate_lunch_std = 0
associate_lunch_free = 0
associate_lunch = df_associate['lunch'].tolist()
for i in associate_lunch :
    if ( i == 'free/reduced'):
        associate_lunch_free += 1
    else:
        associate_lunch_std += 1
    
associate_lunch_total = (associate_lunch_free + associate_lunch_std)/100

bachelors_lunch_std = 0
bachelors_lunch_free = 0
bachelors_lunch = df_bachelors['lunch'].tolist()
for i in bachelors_lunch :
    if ( i == 'free/reduced'):
        bachelors_lunch_free += 1
    else:
        bachelors_lunch_std += 1
    
bachelors_lunch_total = (bachelors_lunch_free + bachelors_lunch_std)/100

masters_lunch_std = 0
masters_lunch_free = 0
masters_lunch = df_masters['lunch'].tolist()
for i in masters_lunch :
    if ( i == 'free/reduced'):
        masters_lunch_free += 1
    else:
        masters_lunch_std += 1
    
masters_lunch_total = (masters_lunch_free + masters_lunch_std)/100

x_indexes = np.arange(5)
width4 = 0.1

plt.rcParams["axes.edgecolor"] = "0.15"
plt.rcParams["axes.linewidth"]  = '1.25'

free_lunch = [high_lunch_free/high_lunch_total,college_lunch_free/college_lunch_total,associate_lunch_free/associate_lunch_total,bachelors_lunch_free/bachelors_lunch_total,masters_lunch_free/masters_lunch_total ]
std_lunch = [high_lunch_std/high_lunch_total,college_lunch_std/college_lunch_total,associate_lunch_std/associate_lunch_total,bachelors_lunch_std/bachelors_lunch_total,masters_lunch_std/masters_lunch_total]
x_dx = ['High School', 'College', 'Associate', 'Bachelors', 'Masters']

plt.bar(x_indexes - width4, free_lunch, width = 0.1, label = 'Free/Reduced')
plt.bar(x_indexes, std_lunch, width = 0.1, label = 'Standard')
plt.xlabel('Parental Level of Eductaion')
plt.ylabel('% of students')
plt.title('Relationship between Parental level of Education & Students Lunch')
plt.xticks(ticks = x_indexes, labels = x_dx)
plt.legend()
plt.grid(True)
plt.style.use('seaborn-poster')
plt.plot()
```

```python
low_level = df_high
medium_level = df_college.append(df_associate, ignore_index = True)
inter_level = medium_level.append(df_bachelors, ignore_index = True)
high_level = df_masters

no_of_category_A = 0
marks_of_category_A = 0
no_of_category_B = 0
marks_of_category_B = 0
no_of_category_C = 0
marks_of_category_C = 0

high_level_lunch = high_level['lunch'].tolist()
inter_level_lunch = inter_level['lunch'].tolist()
low_level_lunch = low_level['lunch'].tolist()

count = 0
for i in (high_level_lunch):
    if ( i == 'standard'):
        no_of_category_A += 1
        marks_of_category_A += high_level.iloc[count, 8]
        count += 1
    else:
        no_of_category_B += 1
        marks_of_category_B += high_level.iloc[count, 8]
        count += 1

count = 0
for i in (inter_level_lunch):
    if ( i == 'standard'):
        no_of_category_A += 1
        marks_of_category_A += inter_level.iloc[count, 8]
        count += 1
    else:
        no_of_category_C += 1
        marks_of_category_C += inter_level.iloc[count, 8]
        count += 1
        
count = 0
for i in (low_level_lunch):
    if ( i == 'standard'):
        no_of_category_B += 1
        marks_of_category_B += low_level.iloc[count, 8]
        count += 1
    else:
        no_of_category_C += 1
        marks_of_category_C += low_level.iloc[count, 8]
        count += 1
        
x_ex = ['Category A', 'Category B', "Category C"]
y_axis = [marks_of_category_A/no_of_category_A, marks_of_category_B/no_of_category_B, marks_of_category_C/no_of_category_C]

plt.rcParams["axes.edgecolor"] = "0.15"
plt.rcParams["axes.linewidth"]  = '1.25'
plt.bar(x_ex, y_axis, width = 0.3 )
plt.style.use('seaborn-poster')
plt.grid(True)
plt.title('Average Marks of Students Category Wise')
plt.ylabel('Average marks of students')
plt.xlabel('Categories')

plt.plot()

'''We divided the students according to their parent’s educational background. Students whose parents 
had a master’s degree, were considered to be highly educated, those with a bachelor’s degree, associate’s
or a college education were considered to have an intermediate level of education and the students whose 
parents had a high school education, were considered to have a low amount of education. Accordingly the 
students were divided into 3 categories, Educational Background: High, Intermediate, and Low.

We defined another parameter “Household Situation” where if, the students had a Standard Lunch and
High/Intermediate Level of Educational Background, they were classified into a ‘Best’ Household situation.
Students who had Free/Reduced Lunch and High Level of Education, or students with a low level of education
but a Standard Lunch were classified as ‘Good’. The students with a Free/Reduced Lunch and having 
Low/Intermediate level of Educational Background were categorized as ‘Okay’.(Type of Lunch also indicates
the Economic Background)'''

'''The results clearly show that the students with a better family background managed to score better 
marks. '''
```

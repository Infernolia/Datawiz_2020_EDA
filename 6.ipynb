```python
import pandas as pd
import matplotlib.pyplot as plt

practice=pd.read_csv('Practice.csv')
```

```python
ax=practice['test preparation course'].value_counts().plot(kind='bar',figsize=(6,6))
ax.set_ylabel('Number of students')
```

```python
ax=practice['gender'].value_counts().plot(kind='bar',figsize=(6,6))
ax.set_ylabel('Number of students')
```

```python
ax=practice['test preparation course'].value_counts().plot(kind='bar',figsize=(8,8))
ax.set_ylabel('Number of students')
```

```python
practice['race/ethnicity'].value_counts().plot(kind='pie',figsize=(11,11))
```

```python
practice['parental level of education'].value_counts().plot(kind='pie',figsize=(11,11))
```

```python
ax=practice['math score'].plot(kind='density',figsize=(14,6))
ax.axvline(practice['math score'].mean(), color='orange')
```

```python
practice['Total score']=practice['math score']+practice['reading score']+practice['writing score']
practice['Average score']=practice['Total score']/3
```

```python
practice.head()
```

```python
practice.plot(kind='scatter',x='Total score',y='math score',figsize=(6,6))
```

```python
densityplot=['math score','reading score','writing score','Average score']
practice[densityplot].plot(kind='density',subplots=True,layout=(4,2), figsize=(22,10))
```

```python
ax=practice[['Average score','race/ethnicity']].boxplot(by='race/ethnicity',figsize=(10,7))
ax.set_ylabel('Average score')
```
